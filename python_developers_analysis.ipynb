```python
import pandas as pd

users_df = pd.read_csv("survey_results_public.csv")

users_df
```

```python
count_respondents = users_df["ResponseId"].count()
print (f"1. Кількість респондентів, які пройшли опитування: {count_respondents}")
```

```python
all_question = users_df.dropna(how="any")
print(f"2. Кількість респондентів, які відповіли на всі запитання: {len(all_question)}")
```

```python
mean_WorkExp = users_df['WorkExp'].mean()
median_WorkExp = users_df['WorkExp'].median()
mode_WorkExp = users_df['WorkExp'].mode()
print (f"3. Значення мір центральної тенденції для досвіду респондентів - середнє: {mean_WorkExp}; медіана: {median_WorkExp}; мода: {mode_WorkExp}")
```

```python
usremotework = 0
for i in users_df.RemoteWork:
    if i == "Remote":
        usremotework = usremotework+1
print(f"4. Кількість респондентів, які працюють віддалено: {usremotework}")
```

```python
python_count = 0
only_inform_lang = users_df.LanguageHaveWorkedWith.dropna()
for response in only_inform_lang:
    language_code = response.split(";")
    if "Python" in language_code:
        python_count += 1
total_respondents = len(users_df) 
python_percentage = round((python_count / total_respondents) * 100, 2)

print (f"5. Відсоток респондентів, що програмують на Python: {python_percentage}")
```

```python
respons_lern_online = users_df.LearnCodeOnline.dropna().count()
print(f"6. Кількість респондентів, які вчились програмувати за допомогою онлайн курсів: {respons_lern_online}")
```

```python
inf_for_analize = users_df[["LanguageHaveWorkedWith", "Country", "ConvertedCompYearly"]]
new_table = inf_for_analize.dropna()
filtred_df = new_table[new_table['LanguageHaveWorkedWith'].str.contains(r'\bPython\b', regex=True)]
grouped = filtred_df.groupby("Country")
result_yearly = grouped["ConvertedCompYearly"].agg(["mean", "median"])
print("7. Cередня та медіанна сума компенсації (ConvertedCompYearly) в кожній країні")
result_yearly
```

```python
education_converted = users_df[["ConvertedCompYearly", "EdLevel"]]
filtr_education_converted = education_converted.dropna()
sorted_filtr_education_converted = filtr_education_converted.sort_values(by = "ConvertedCompYearly", ascending = False)
print("8. Pівні освіти мають 5 респондентів з найбільшою компенсацією")
sorted_filtr_education_converted[:5]
```
